```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

# IPython magic to create plots within cells and other vars for looks
%matplotlib inline
pd.options.display.max_rows = 200
matplotlib.style.use('ggplot')
```

```python
frbh41 ='https://www.federalreserve.gov/datadownload/Output.aspx?rel=H41&series=e6abcc59dd71a2c3427d7862d2df2715&lastobs=100&from=&to=&filetype=csv&label=include&layout=seriescolumn'
```

```python
# CSV generated from URL stored in frbh41 above
# Just re-run the whole notebook to get the latest data

#fed = pd.read_csv('FRB_H41.csv', parse_dates=['Series Description'], skiprows=[1,2,3,4,5])
fed = pd.read_csv(frbh41, parse_dates=['Series Description'], skiprows=[1,2,3,4,5])

fed['ust'] = fed['Assets: Securities Held Outright: U.S. Treasury securities: Wednesday level']
fed = fed.set_index('Series Description')
fed.ust[5:].astype(int).plot(title='FED RESERVE US Treasury Securities')
```
